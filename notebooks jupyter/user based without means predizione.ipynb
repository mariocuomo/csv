## SPLIT DATI TRAIN E TEST

```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(df, test_size = 0.2, random_state = 42)
```

## CREAZIONE RATING MATRIX

```python
user_data = X_train.pivot(index = 'userId', columns = 'movieId', values = 'rating').fillna(0)
```

```python
user_data.head()
```

```python
user_data.tail()
```

## COPIA TRAIN E TEST

```python
dummy_train = X_train.copy()
dummy_test = X_test.copy()
```

## FILM NON REATED DALL'UTENTE SONO TAGGATI CON 0 SUL DUMMY_TRAIN, CON 1 SUL DUMMY_TEST

```python
dummy_train['rating'] = dummy_train['rating'].apply(lambda x: 0 if x > 0 else 1)
```

```python
dummy_train = dummy_train.pivot(index = 'userId', columns = 'movieId', values = 'rating').fillna(1)

dummy_test = dummy_test.pivot(index ='userId', columns = 'movieId', values = 'rating').fillna(0)
```

## COSENO SIMILARITA'

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_data)
user_similarity[np.isnan(user_similarity)] = 0
print(user_similarity)
print(user_similarity.shape)
```

```python
user_predicted_ratings = np.dot(user_similarity, user_data)
user_predicted_ratings
```

## SONO STATI PREDETTI TUTTI I FILM, ANCHE QUELLI CHE L'UTENTE HA GIA VISIONATO

```python
user_final_ratings = np.multiply(user_predicted_ratings, dummy_train)
user_final_ratings.head()
```

## SI PREDICONO I PRIMI 30 FILM CHE POTREBBERO PIACERE ALL'UTENTE 1

```python
ls = user_final_ratings.iloc[user-1].sort_values(ascending = False)[0:100]
_ls=[]

for item in ls.iteritems():
    _ls.append(item)

print(_ls[0])
```

```python
ll = []
for item in _ls:
    title=df_film.loc[df_film['movieId'] == item[0]].iloc[0]['title']
    score = item[1]
    ll.append([title, score])
    
suggestions = pd.DataFrame(ll, columns = ['Movie', 'Score'])


print("****TOP 30 SUGGESTIONS FOR YOU:****\n")
suggestions.head(30)
```
