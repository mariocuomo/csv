```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

### CARICO IL DATASET IN MEMORIA


```python
user=276
df = pd.read_csv('C:\\Users\\cuomo\\Desktop\\tesina sistemi intelligenti per internet\\datasets\\ratings.csv')
df_film = pd.read_csv('C:\\Users\\cuomo\\Desktop\\tesina sistemi intelligenti per internet\\datasets\\movies.csv')
```

```python
rating_matrix = df.pivot(index = 'userId', columns = 'movieId', values = 'rating').fillna(0)
```

```python
rating_matrix
```

### SOTTRAI LA MEDIA DELL'UTENTE A OGNI SUO RATE - ADJUSTED COSINE SIMILARITY


```python
new = rating_matrix.sub(rating_matrix.mean(axis=1), axis=0)
```

```python
item_similarity = cosine_similarity(new.transpose())
item_similarity
```

```python
item_similarity[np.isnan(item_similarity)] = 0
dd=pd.DataFrame(item_similarity)
dd.columns = rating_matrix.columns
dd.insert(0, 'movieId', dd.columns)
dd
```

```python
dd
```

```python
#crea lista di film visionati dall'utente
film_rating = []
for index, value in rating_matrix.iloc[user-1].items():
    film_rating.append((index,value))
film_rating
```

```python
film_rating
```

```python
film_rated = list(filter(lambda item: item[1] != 0, film_rating))
film_rated
```

```python
filmVistiEListaSimili=[]
for (film, rate) in film_rated:
    filmVistiEListaSimili.append((film, rate, dict(list(((dd[dd['movieId'] == film]).squeeze()).items()))))
filmVistiEListaSimili[0]
```

```python
#per ogni film non visionato dall'utente predici un rating sulla base dei rating dati
rating_stimato=[]
for film,_ in film_rated:
    den=0
    num=0
    for (_film, rate, dictSimili) in filmVistiEListaSimili:
        if(film!=_film):
            sim = dictSimili[film]
            num=num+(rate*sim)
            den=den+sim
    if(den==0):
        rating_stimato.append((film,0))
    else:
        rating_stimato.append((film,num/den))
```

```python
ll = []
for item in rating_stimato:
    rslt_df = df_film[df_film['movieId'] == (item[0])]
    title=rslt_df.iloc[0]['title']
    genere=rslt_df.iloc[0]['genres'].replace("|", ",")
    score = item[1]
    ll.append([title, score, genere])

suggestions = pd.DataFrame(ll, columns = ['TITOLO', 'RATING PREDETTO', 'GENERE'])
```
