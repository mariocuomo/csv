### LIBRERIE

```python
import pandas as pd
import numpy as np
import scipy.stats
```

### CARICO IL DATASET IN MEMORIA

```python
df = pd.read_csv('https://raw.githubusercontent.com/mariocuomo/progettoSII/main/ratings.csv')
df_film = pd.read_csv('https://raw.githubusercontent.com/mariocuomo/progettoSII/main/movies.csv')
```

```python
### CREO RATING MATRIX
```

```python
rating_matrix = df.pivot(index = 'userId', columns = 'movieId', values = 'rating').fillna(0)
```

```python
rating_matrix
```

### PER OGNI COLONNA (film) CALCOLO L'ENTROPIA

```python
lst=[]
for (colname,colval) in rating_matrix.iteritems():
    lst.append((colname,scipy.stats.entropy(colval)))
lst.sort(key = lambda x: x[1], reverse=True)
```

```python
lst[2]
```

### ARRICCHISCO LA STAMPA CON LE INFORMAZIONI RIGUARDO IL TITOLO E IL GENERE

```python
ll = []
for item in lst[0:50]:
    rslt_df = df_film[df_film['movieId'] == (item[0])]
    title=rslt_df.iloc[0]['title']
    genere=rslt_df.iloc[0]['genres'].replace("|", ",")
    score = item[1]
    ll.append([title, score, genere])
suggestions = pd.DataFrame(ll, columns = ['TITOLO', 'ENTROPIA', 'GENERE'])
```
